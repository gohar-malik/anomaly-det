```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
model_name = "model_ResNet18_cifar100_b128_ep300_newms"
log_file = os.path.join("/home/gohar/DCV/anomaly-det/train_resnet/logs", f"{model_name}.out")
```

```python
with open(log_file, 'r') as f:
    lines = f.readlines()

    epochs = []
    train_loss = []
    test_loss = []
    test_acc = []
    lr = [] 

    for line in lines:
        if "Epoch:" in line:
            epochs.append(int(line.split("Epoch:")[1]))
        
        elif "Training:" in line:
            train_loss.append(float(line.split("Loss:")[1][:7]))
            lr.append(float(line.split("LR:")[1][:9]))
        
        elif "Testing:" in line:
            test_loss.append(float(line.split("Loss:")[1][:7]))
            test_acc.append(float(line.split("Acc:")[1][:7]))

print(len(epochs))
```

```python
start_epoch = 150
plt.plot(epochs[start_epoch:], lr[start_epoch:], marker = "o", label = "train")
# plt.plot(epochs[start_epoch:], test_loss[start_epoch:], marker = "^", label = "val")
plt.grid()
plt.title("Loss Plot")
plt.xlabel("Epochs")
plt.legend()
```

```python
# start_epoch = 150
plt.plot(epochs[start_epoch:], train_loss[start_epoch:], marker = "o", label = "train")
# plt.plot(epochs[start_epoch:], test_loss[start_epoch:], marker = "^", label = "val")
plt.grid()
plt.title("Loss Plot")
plt.xlabel("Epochs")
plt.legend()
```

```python
# start_epoch = 150
plt.plot(epochs[start_epoch:], test_acc[start_epoch:])
plt.grid()
plt.title("Test Acc Plot")
plt.xlabel("Epochs")
```
